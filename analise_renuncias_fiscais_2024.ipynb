Segundo o Portal da Transparência, a renúncia fiscal ocorre quando o governo abre mão de receber o total ou parte dos tributos devidos em prol de um estímulo da economia ou de programas sociais, que serão desenvolvidos pelo setor privado ou por entidades não governamentais.

Assim, o ato de renunciar a uma receita pública é uma política de governo, com o objetivo, dentre outros, de incentivar o desenvolvimento de setores econômicos estratégicos ou de regiões do país. Esse é o caso, por exemplo, da Zona Franca de Manaus.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/projeto_final/2024_RenunciasFiscais.csv', encoding='latin1', sep=';')
```

```python
df.head(5)
```

```python
df['Valor Renúncia Fiscal (R$)'] = ( df['Valor Renúncia Fiscal (R$)'].str.replace(',', '.', regex=False).astype(float))
```

```python
#Valor por Tributo
valor_por_tributo = df.groupby('Tributo')['Valor Renúncia Fiscal (R$)'].sum()
valor_por_tributo = valor_por_tributo.sort_values(ascending=False)
valor_por_tributo
```

```python
#Renúncia por UF
renuncia_por_uf = df.groupby('UF')['Valor Renúncia Fiscal (R$)'].sum()
renuncia_por_uf = renuncia_por_uf.sort_values(ascending=False)
renuncia_por_uf.head(5)
```

```python
#Renúncia de PIS por UF
pis_por_uf = df[df['Tributo'] == 'PIS'].groupby('UF')['Valor Renúncia Fiscal (R$)'].sum()
pis_por_uf = pis_por_uf.sort_values(ascending=False)
pis_por_uf.head(5)
```

```python
#Renúncia de Imposto de Importação por UF
impi_por_uf = df[df['Tributo'] == 'Imposto de Importação'].groupby('UF')['Valor Renúncia Fiscal (R$)'].sum()
impi_por_uf = impi_por_uf.sort_values(ascending=False)
impi_por_uf.head(5)
```

```python
#Renúncia de COFINS por UF
cofins_por_uf = df[df['Tributo'] == 'COFINS'].groupby('UF')['Valor Renúncia Fiscal (R$)'].sum()
cofins_por_uf = cofins_por_uf.sort_values(ascending=False)
cofins_por_uf.head(5)
```

```python
#Renúncia de IPI por UF
ipi_por_uf = df[df['Tributo'] == 'IPI'].groupby('UF')['Valor Renúncia Fiscal (R$)'].sum()
ipi_por_uf = ipi_por_uf.sort_values(ascending=False)
ipi_por_uf.head(5)
```

```python
#Importando Bibliotecas para criar os Gráficos
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Gráfico da Renúncia Fiscal por UF
df_top10_uf = renuncia_por_uf.head(10).reset_index()
df_top10_uf['Valor (Milhões)'] = df_top10_uf['Valor Renúncia Fiscal (R$)'] / 1000000
fig_uf = px.bar(
    df_top10_uf,
    x= 'UF',
    y='Valor (Milhões)',
    title='Top 10 UFs com Maior Renúncia Fiscal',
    text_auto='.2f',
    color= 'Valor (Milhões)',
    color_continuous_scale='YlGn')
fig_uf.update_traces(textposition='outside')
fig_uf.update_yaxes(range=[0, df_top10_uf['Valor (Milhões)'].max() * 1.15])
fig_uf.show()
```

```python
#Gráfico da Renúncia de PIS por UF
df_top10_pis = pis_por_uf.head(10).reset_index()
df_top10_pis['Valor (Milhões)'] = df_top10_pis['Valor Renúncia Fiscal (R$)'] / 1000000
fig_pis = px.bar(
    df_top10_pis,
    x='UF',
    y='Valor (Milhões)',
    title='Top 10 UFs com Maior Renúncia de PIS (em Milhões de R$)',
    text_auto='.2f',
    color='Valor (Milhões)',
    color_continuous_scale='YlGn' )
fig_pis.update_traces(textposition='outside')
fig_pis.update_yaxes(range=[0, df_top10_pis['Valor (Milhões)'].max() * 1.15])
fig_pis.show()
```

```python
#Gráfico da Renúncia de Imposto de Importação por UF
df_top10_impi = impi_por_uf.head(10).reset_index()
df_top10_impi['Valor (Milhões)'] = df_top10_impi['Valor Renúncia Fiscal (R$)'] / 1000000
fig_impi = px.bar(
    df_top10_impi,
    x='UF',
    y='Valor (Milhões)',
    title='Top 10 UFs com Maior Renúncia de Imposto de Importação (em Milhões de R$)',
    text_auto='.2f',
    color='Valor (Milhões)',
    color_continuous_scale='YlGn')
fig_impi.update_traces(textposition='outside')
fig_impi.update_yaxes(range=[0, df_top10_impi['Valor (Milhões)'].max() * 1.15])
fig_impi.show()
```

```python
#Gráfico da Renúncia de COFINS por UF
df_top10_cofins = cofins_por_uf.head(10).reset_index()
df_top10_cofins['Valor (Milhões)'] = df_top10_cofins['Valor Renúncia Fiscal (R$)'] / 1000000
fig_cofins = px.bar(
    df_top10_cofins,
    x='UF',
    y='Valor (Milhões)',
    title='Top 10 UFs com Maior Renúncia de COFINS (em Milhões de R$)',
    text_auto='.2f',
    color= 'Valor (Milhões)',
    color_continuous_scale='YlGn')
fig_cofins.update_traces(textposition='outside')
fig_cofins.update_yaxes(range=[0, df_top10_cofins['Valor (Milhões)'].max() * 1.15])
fig_cofins.show()
```
